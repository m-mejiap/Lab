```python
import numpy as np
import matplotlib.pylab as plt
import time
from scipy.optimize import curve_fit
```

#### _Recursive method._

```python
def fib_recursion(n): 
    if (n == 0):
        n_fibonacci = 0
    if (n == 1):
        n_fibonacci = 1
    if (n >= 2):
        n_fibonacci = fib_recursion(n-1) + fib_recursion(n-2)
    return n_fibonacci

list_of_n = []
list_of_times = []

for i in range(35):
    init=time.time()
    fib_recursion(i)
    difference_time=time.time()-init
    list_of_n.append(i)
    list_of_times.append(difference_time)
```

```python
def linear(x, m, b):
    return (m*x) + b

param1 = curve_fit(linear, list_of_n, list_of_times)[0]   

def polynomial2(x, a, b, c):
    return a + b*x + c*(x**2)

param2 = curve_fit(polynomial2, list_of_n, list_of_times)[0]

def polynomial3(x, a, b, c, d):
    return a + b*x + c*(x**2) + d*(x**3)

param3 = curve_fit(polynomial3, list_of_n, list_of_times)[0]

def polynomial4(x, a, b, c, d, e):
    return a + b*x + c*(x**2) + d*(x**3) + e*(x**4)

param4 = curve_fit(polynomial4, list_of_n, list_of_times)[0]

def polynomial5(x, a, b, c, d, e, f):
    return a + b*x + c*(x**2) + d*(x**3) + e*(x**4) + f*(x**5)

param5 = curve_fit(polynomial5, list_of_n, list_of_times)[0]

def polynomial6(x, a, b, c, d, e, f, g):
    return a + b*x + c*(x**2) + d*(x**3) + e*(x**4) + f*(x**5) + g*(x**6)

param6 = curve_fit(polynomial6, list_of_n, list_of_times)[0]

def polynomial7(x, a, b, c, d, e, f, g, h):
    return a + b*x + c*(x**2) + d*(x**3) + e*(x**4) + f*(x**5) + g*(x**6) + h*(x**7)

param7 = curve_fit(polynomial7, list_of_n, list_of_times)[0]

def polynomial8(x, a, b, c, d, e, f, g, h, i):
    return a + b*x + c*(x**2) + d*(x**3) + e*(x**4) + f*(x**5) + g*(x**6) + h*(x**7) + i*(x**8)

param8 = curve_fit(polynomial8, list_of_n, list_of_times)[0]

def polynomial9(x, a, b, c, d, e, f, g, h, i, j):
    return a + b*x + c*(x**2) + d*(x**3) + e*(x**4) + f*(x**5) + g*(x**6) + h*(x**7) + i*(x**8) + j*(x**9)

param9 = curve_fit(polynomial9, list_of_n, list_of_times)[0]

def polynomial10(x, a, b, c, d, e, f, g, h, i, j, k):
    return a + b*x + c*(x**2) + d*(x**3) + e*(x**4) + f*(x**5) + g*(x**6) + h*(x**7) + i*(x**8) + j*(x**9) + k*(x**10)

param10 = curve_fit(polynomial10, list_of_n, list_of_times)[0]

def exponential(x, a, b, c):
    return a * np.exp(-b * x) + c

param11 = curve_fit(exponential, list_of_n, list_of_times, maxfev=5000)[0]
```

```python
fig = plt.figure(figsize=(30,20))

z = fig.add_subplot(3,4,1)
z.scatter(list_of_n,list_of_times, color="firebrick",marker='.')
z.set_xlabel("n.")
z.set_ylabel("Time.")
z.set_title("Time that Fibonacci's recursive method takes to perform in function of the size n.")
            
a = fig.add_subplot(3,4,2)
a.scatter(list_of_n,linear(np.array(list_of_n),*param1), color="black",marker='.')
a.set_xlabel(u"n.")
a.set_ylabel(u"Time.")
a.set_title(u"Linear fit.")

b = fig.add_subplot(3,4,3)
b.scatter(list_of_n,polynomial2(np.array(list_of_n),*param2), color="black",marker='.')
b.set_xlabel(u"n.")
b.set_ylabel(u"Time.")
b.set_title(u"2th degree polynomial fit.")

c = fig.add_subplot(3,4,4)
c.scatter(list_of_n,polynomial3(np.array(list_of_n),*param3), color="black",marker='.')
c.set_xlabel(u"n.")
c.set_ylabel(u"Time.")
c.set_title(u"3th degree polynomial fit.")

d = fig.add_subplot(3,4,5)
d.scatter(list_of_n,polynomial4(np.array(list_of_n),*param4), color="black",marker='.')
d.set_xlabel(u"n.")
d.set_ylabel(u"Time.")
d.set_title(u"4th degree polynomial fit.")

e = fig.add_subplot(3,4,6)
e.scatter(list_of_n,polynomial5(np.array(list_of_n),*param5), color="black",marker='.')
e.set_xlabel(u"n.")
e.set_ylabel(u"Time.")
e.set_title(u"5th degree polynomial fit.")

f = fig.add_subplot(3,4,7)
f.scatter(list_of_n,polynomial6(np.array(list_of_n),*param6), color="black",marker='.')
f.set_xlabel(u"n.")
f.set_ylabel(u"Time.")
f.set_title(u"6th degree polynomial fit.")

g = fig.add_subplot(3,4,8)
g.scatter(list_of_n,polynomial7(np.array(list_of_n),*param7), color="black",marker='.')
g.set_xlabel(u"n.")
g.set_ylabel(u"Time.")
g.set_title(u"7th degree polynomial fit.")

h = fig.add_subplot(3,4,9)
h.scatter(list_of_n,polynomial8(np.array(list_of_n),*param8), color="black",marker='.')
h.set_xlabel(u"n.")
h.set_ylabel(u"Time.")
h.set_title(u"8th degree polynomial fit.")

i = fig.add_subplot(3,4,10)
i.scatter(list_of_n,polynomial9(np.array(list_of_n),*param9), color="black",marker='.')
i.set_xlabel(u"n.")
i.set_ylabel(u"Time.")
i.set_title(u"9th degree polynomial fit.")

j = fig.add_subplot(3,4,11)
j.scatter(list_of_n,polynomial10(np.array(list_of_n),*param10), color="black",marker='.')
j.set_xlabel(u"n.")
j.set_ylabel(u"Time.")
j.set_title(u"10th degree polynomial fit.")

k = fig.add_subplot(3,4,12)
k.scatter(list_of_n,exponential(np.array(list_of_n),*param11), color="black",marker='.')
k.set_xlabel(u"n.")
k.set_ylabel(u"Time.")
k.set_title(u"Exponential fit.")

plt.show()
```

```python
def R(yp,y):
    a = np.var(yp - y)
    b = np.var(y)
    return 1 - (a/b)

print("Linear fit: R =", R(linear(np.array(list_of_n),*param1),list_of_times))
print("2th degree polynomial fit: R =", R(polynomial2(np.array(list_of_n),*param2),list_of_times))
print("3th degree polynomial fit: R =", R(polynomial3(np.array(list_of_n),*param3),list_of_times))
print("4th degree polynomial fit: R =", R(polynomial4(np.array(list_of_n),*param4),list_of_times))
print("5th degree polynomial fit: R =", R(polynomial5(np.array(list_of_n),*param5),list_of_times))
print("6th degree polynomial fit: R =", R(polynomial6(np.array(list_of_n),*param6),list_of_times))
print("7th degree polynomial fit: R =", R(polynomial7(np.array(list_of_n),*param7),list_of_times))
print("8th degree polynomial fit: R =", R(polynomial8(np.array(list_of_n),*param8),list_of_times))
print("9th degree polynomial fit: R =", R(polynomial9(np.array(list_of_n),*param9),list_of_times))
print("10th degree polynomial fit: R =", R(polynomial10(np.array(list_of_n),*param10),list_of_times))
print("Exponential fit: R =", R(exponential(np.array(list_of_n),*param11),list_of_times))
```

What is the best fit according to the result of R? According to R, the best fit is a 10th degreee polynomial fit.

```python
y_minus = list_of_n - np.var(list_of_n)
y_plus = list_of_n + np.var(list_of_n)
plt.plot(list_of_n, polynomial10(np.array(list_of_n),*param10), 'k-')
plt.fill_between(list_of_n, y_minus, y_plus)
plt.show()
```

#### _Linear method._

```python
def fib_optimized(n):
    a = 1
    b = 0
    i = 0
    if(n == 0):
        n_fibonacci = 0
    else:
        while i < n:
            n_fibonacci = a
            a,b = a+b,a
            i = i + 1
    return n_fibonacci

list_of_n2 = []
list_of_times2 = []

for i in range(100):
    init=time.time()
    fib_optimized(i)
    difference_time=time.time()-init
    list_of_n2.append(i)
    list_of_times2.append(difference_time)
```

## 2. Fitting Gaussians.

```python
def Gausian_2D(x,y,mu_x,mu_y,sigma_x,sigma_y):
    return 1.0/(2*np.pi*sigma_x*sigma_y)*np.exp(-0.5*((x-mu_x)**2/sigma_x**2)-0.5*((y-mu_y)**2/sigma_y**2))

x=np.linspace(-5,5,1000)
y=np.linspace(-8,8,1000)
X,Y=np.meshgrid(x,y)
z=Gausian_2D(X,Y,0,0,1,3)

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,z,cmap=plt.cm.coolwarm,linewidth=0,antialiased=False)
plt.show()
```

```python
WhiteNoise = np.random.normal(0,0.01,1000)
znoise = z + WhiteNoise

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,znoise,cmap=plt.cm.coolwarm,linewidth=0,antialiased=False)
plt.show()
```

```python
data = np.array([x,y])

def Gausian_DosD(data0,mu_x,mu_y,sigma_x,sigma_y):
    return 1.0/(2*np.pi*sigma_x*sigma_y)*np.exp(-0.5*((data0[0]-mu_x)**2/sigma_x**2)-0.5*((data0[1]-mu_y)**2/sigma_y**2))

z2 = Gausian_DosD(data,0,0,1,3)
z2noise = z2 + WhiteNoise
parameters = curve_fit(Gausian_DosD, data, z2noise)[0]
```

```python
print(u"1st parameter: Introduced = 0, Fit =",  parameters[0], ", Difference =", np.abs(parameters[0]-0))
print(u"2nd parameter: Introduced = 0, Fit =",  parameters[1], ", Difference =", np.abs(parameters[1]-0))
print(u"3rd parameter: Introduced = 1, Fit =",  parameters[2], ", Difference =", np.abs(parameters[2]-1))
print(u"4th parameter: Introduced = 3, Fit =",  parameters[3], ", Difference =", np.abs(parameters[3]-3))
```

The parameters obtained with the fit make sense according to the real (introduced) parameters. We can realize that the difference between the introduced and the fit parameters is always a value less than 1. Therefore, we can conclude that fit parameters are close from what we expected.

```python
parameters2 = curve_fit(Gausian_DosD, data, z2noise, p0=[0,0,1,3])[0]

print(u"1st parameter: Introduced = 0, Fit =",  parameters2[0], ", Difference =", np.abs(parameters2[0]-0))
print(u"2nd parameter: Introduced = 0, Fit =",  parameters2[1], ", Difference =", np.abs(parameters2[1]-0))
print(u"3rd parameter: Introduced = 1, Fit =",  parameters2[2], ", Difference =", np.abs(parameters2[2]-1))
print(u"4th parameter: Introduced = 3, Fit =",  parameters2[3], ", Difference =", np.abs(parameters2[3]-3))
```

If we refit the curve giving some initial conditions by using the parameter p0, the results improve considerably as we can see in each parameter difference.

```python
#Plot using the fit parameters.
Z2 = Gausian_2D(X,Y,*parameters)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z2,cmap=plt.cm.coolwarm,linewidth=0,antialiased=False)
plt.show()
```
